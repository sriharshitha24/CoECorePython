```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title("Simple Line plot")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y,color='red',linestyle='dashed',linewidth=2,marker='*')#can be done by using multiple attrinutes together
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
#plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['Jan','Feb','March','April','May','June']
y=[5000,7500,8000,9000,12000,20000]
plt.plot(x,y,color='green',linestyle='dashed',linewidth=2,marker='*')
plt.title("Growth of Savings account")
plt.xlabel("Month")
plt.ylabel("Savings")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['skyblue','red','blue','yellow','green'],edgecolor='black',)
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35,36,40,42,49,49,48,50,52,57,53,58,59,53,51,59,60,61,67,65,64,67,68,69,68,70]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor='black')
plt.title("Age Distribution of Customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
#plt.grid(True)
plt.show()
```

```python
a=['python','ruby','C++','Java',]
b=[40,10,20,30]
colors=['skyblue','blue','red','yellow']
plt.pie(b,labels=a,colors=['skyblue','yellow','red','green'],autopct='%1.1f%%')
plt.legend()
plt.title("Programming Language Usage")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title("Scatter plot")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.show()
```
